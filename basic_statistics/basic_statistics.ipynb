## Breif Interpretation

#### Mean - The mean tells you the central value of the column. It represents the overall average.
#### Median - The median tells you the middle point of the data when sorted.It is not affected by outliers, so it shows the true central tendency better when data is skewed.
#### Mode - he mode tells you which value appears most often.Useful for understanding common patterns or repeated values.
#### Standard Deviation - The standard deviation (std) tells you how spread out the values are.
#### Low std → values are close to the mean (less variation).
#### High std → values vary a lot (more spread).


## Data Vizualization 

#### 	Plot histograms for each numerical column. Analyze the distribution (e.g., skewness, presence of outliers) and provide inferences.


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt

# Function to create styled histogram
def plot_hist(column):
    plt.figure(figsize=(6,3))
    plt.title(f'{column} Distribution')
    plt.ylabel('Frequency')
    plt.xlabel(column)

    plt.hist(df[column],
             color='#FF2056',
             edgecolor='white',
             bins=10
            )
    plt.show()


# Plot for each numerical column
num_cols = ['Volume', 'Avg Price', 'Total Sales Value', 
            'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']

for col in num_cols:
    plot_hist(col)
```

### Inferences

#### Volume Distribution:
##### Mostly low values (1–10).

##### Very few high-volume rows.

##### Long tail on the right → right-skewed.

##### Inferences:

##### Most products sell in small quantities.

##### A few products with very high volume may be outliers.

##### Sales are dominated by low-volume items.


#### Avg Price

##### Inferences:
