```python
patient_df_cleaned =df.drop(
    columns=['ReceptionistID', 'CheckInTime'],inplace=True)
patient_df_cleaned  
```

```python
dept_total_bill = df.groupby('Department')['BillAmount'].sum()
print(dept_total_bill)
```

```python
patient_df_cleaned = df.drop_duplicates(subset='PatientID')
patient_df_cleaned
```

```python
mean_bill = df['BillAmount'].mean()
patient_df_cleaned['BillAmount'] = patient_df_cleaned['BillAmount'].fillna(mean_bill)
patient_df_cleaned
```

```python
billing_data=pd.read_csv('Billing_data.csv')
```

```python
merged_df = pd.merge(
    patient_df_cleaned,
    billing_data,
    on='PatientID',
    how='left',
    suffixes=('', '_billing')
)
merged_df
```

```python
new_patients = pd.DataFrame({
    "PatientID": [106, 107, 108, 109, 110],
    "Name": ["Rakshitha", "Rahul Verma", "Sneha Iyer", "Kiran Patil", "Neha Singh"],
    "Department": ["Cardiology", "Orthopedics", "Neurology", "Pediatrics", "Dermatology"],
    "Doctor": ["Dr. Mehta", "Dr. Sharma", "Dr. Kulkarni", "Dr. Desai", "Dr. Malhotra"],
    "BillAmount": [15000, 22000, 30000, 8000, 6000]
})

merged_df = pd.concat([merged_df, new_patients], axis=0)
merged_df
```

```python
# 1. Fill missing InsuranceCovered with default value
final_df = merged_df.copy()
final_df["InsuranceCovered"] = final_df["InsuranceCovered"].fillna(2000)

# 2. Recalculate FinalAmount safely
final_df["FinalAmount"] = final_df["BillAmount"] - final_df["InsuranceCovered"]
```
